Создание дерева по принципу ID3 (ветвей больше 2)

```python
import pandas as pd
from math import log2

# Функция для рассчета энтропии
def entropy(counts):
    total = sum(counts)
    return -sum([ (count/total) * log2(count/total) if count else 0
                 for count in counts])

# Функция для рассчета прироста информации
def information_gain(data, feature, target):
    total_entropy = entropy(data[target].value_counts().tolist())

    weighted_entropy = 0
    for value in data[feature].unique():
        subset = data[data[feature] == value]
        weighted_entropy += (len(subset) / len(data)) * entropy(subset[target].value_counts().tolist())

    return total_entropy - weighted_entropy

# Функция для построения дерева решений
def build_decision_tree(data, features, target):
    # Базовый случай: все примеры принадлежат одному классу
    if len(data[target].unique()) == 1:
        return data[target].iloc[0]

    # Базовый случай: нет больше признаков для разделения
    if len(features) == 0:
        return data[target].value_counts().idxmax()

    # Выбор лучшего признака для разделения
    best_feature = max(features, key=lambda feature: information_gain(data, feature, target))

    # Создание узла дерева
    tree = {best_feature: {}}

    # Удаление выбранного признака из списка
    remaining_features = [f for f in features if f != best_feature]

    # Рекурсивное построение поддеревьев
    for value in data[best_feature].unique():
        subset = data[data[best_feature] == value]
        subtree = build_decision_tree(subset, remaining_features, target)
        tree[best_feature][value] = subtree

    return tree

# Построение дерева
features = ['Формат', 'Продолжительность', 'Аудитория', 'Наличие билетов', 'Интерактивность', 'Цель', 'Место проведения']
tree = build_decision_tree(df.copy(), features, 'Класс')

# Вывод дерева
import pprint
pprint.pprint(tree)
```

Бинарное дерево классификаций - использование готовой библиотеки

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz

# Создание и обучение модели
model = DecisionTreeClassifier()
X = df.drop('Класс', axis=1)
y = df['Класс']
X_enc = pd.get_dummies(X, drop_first= True)
model.fit(X_enc, y)

# Визуализация дерева
def visualize_tree(tree_model, feature_names):
    tree_data = export_graphviz(tree_model,
                                feature_names=feature_names,
                                class_names=sorted(y.unique()),max_depth=3,
                                out_file=None,
                                filled=True,
                                rounded=True,
                                special_characters=True,
                                impurity=True,
                                label='all',  # Добавляем вывод класса и критерия
                               )
    graph = graphviz.Source(tree_data)
    return graph
# Вызов функции визуализации
graph = visualize_tree(model, feature_names=X_enc.columns) # Передаем объект модели
graph.render("decision_tree")
graph
```
