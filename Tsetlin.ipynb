```python
import matplotlib.pyplot as plt
import numpy as np
```

# Задание 1 Равенство развития территорий S=P

```python
task = "Задание 1 Равенство развития территорий $S=P$"
x = np.arange(0, 101)
S = 400 * x - 0.02 * (x**3)
P = 24000 - 200 * x - 0.4 * x**2
```

```python
plt.plot(x, S, label="S")
plt.plot(x, P, label="P")
plt.legend()
plt.xlim((0, 100))
plt.ylim(0)
plt.axvline(x=41, color="k")
plt.title(task)
plt.savefig(f"{task}.png")
```

```python
m = 1e9
a = 0
for i in range(S.shape[0]):
    if m > abs(S[i] - P[i]):
        a = x[i]
        m = abs(S[i] - P[i])
m, a
```
